```python
drive.mount('/content/drive')
case_df = pd.read_csv('/content/drive/MyDrive/KDT_240424/m9_LLM/data/case_df(101,13,6).csv')

case_df.head()
```

```python
training_data = []

system_message = "You are a car accident lawyer with 20 years of experience. Your task is to determine the ratio of negligence for each case provided\
 and to clearly explain the reasoning behind your assessment based on the details of each case."

def create_user_message(row):
    return row['사고내용']

def prepare_example_conversation(row):
    messages = []
    messages.append({"role": "system", "content": system_message})

    user_message = create_user_message(row)
    messages.append({"role": "user", "content": user_message})

    messages.append({"role": "assistant", "content":
            f"""
            과실 비율: {row['과실비율']}\n\n
            사고 상황: {row['사고상황']}\n\n
            과실 해설: {row['과실해설']}\n\n
            관련 법규: {row['관련법규']}\n\n
            판결 사례: {row['판결사례']}\n\n
            애니메이션 : {row['애니메이션']}
            """})

    return {"messages": messages}

pprint(prepare_example_conversation(case_df.iloc[0]))
```

```python
# use the first 100 rows of the dataset for training
training_df = case_df.loc[0:100]

# apply the prepare_example_conversation function to each row of the trainging_df
training_data = training_df.apply(prepare_example_conversation, axis=1).tolist()

for example in training_data[:5]:
    print(example)
```

```python
validation_df = case_df.loc[101:121]
validation_data = validation_df.apply(prepare_example_conversation, axis=1).tolist()
print(validation_df.head(2))
print(validation_df.tail(2))
```

그런 다음 각 줄이 하나의 훈련 대화 사례가 되도록 .jsonl 파일 데이터로 저장을 해야합니다
- 대규모 데이터를 처리할 때 유용하며, 한 줄씩 데이터를 읽거나 쓸 수 있기 때문에 효율적입니다

```python
def write_jsonl(data_list: list, filename: str) -> None:
    with open(filename, 'w', encoding='utf-8') as out:
        for ddict in data_list:
            jout = json.dumps(ddict, ensure_ascii=False) + '\n'
            out.write(jout)
```

```python
training_file_name = "tmp_accident_finetune_training.jsonl"
write_jsonl(training_data, training_file_name)

validation_file_name = "tmp_accident_finetune_validation.jsonl"
write_jsonl(validation_data, validation_file_name)
```

#Fine-Tuning

```python
from openai import OpenAI
client = OpenAI(api_key='')
MODEL = "gpt-4o-mini-2024-07-18"

train = client.files.create(
    file=open("tmp_accident_finetune_training.jsonl", "rb"),
    purpose='fine-tune'
)

validation = client.files.create(
    file=open("tmp_accident_finetune_validation.jsonl", "rb"),
    purpose='fine-tune'
)
```

```python
train
```

```python
validation
```

```python
client.fine_tuning.jobs.create(
    training_file=train.id,
    validation_file=validation.id,
    model=MODEL,
    hyperparameters={
        "n_epochs": 5,
    },
)
```

```python
# 작업 상태 확인
response = client.fine_tuning.jobs.retrieve('ftjob-ANa0i9Zjan2Jrc09VauBTPd4')

# 작업 상태 학인
status = response.status

if status == 'succeeded':
    print('작업이 완료되었습니다')
    # 작업이 완료되면 결과를 가져와서 사용 할 수 있습니다
    # 예를 들어, 모델을 사용하여 텍스트 생성 또는 다른 작업을 수행 할 수 있습니다
else:
    print(f"작업상태 : {status}")

total_tokens = response.trained_tokens
print(f"사용 된 총 토큰 수 : {total_tokens}")

```
